```python
y = df['prognosis']
X = df[l1] 
labels = df['prognosis']

# label encode the target variable
y = LabelEncoder().fit_transform(y)

# Create dictionary from arrays
my_dict = {key: value for key, value in zip(labels, y)}

# Print the resulting dictionary
print(my_dict)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
clf1 = tree.DecisionTreeClassifier(max_depth=37)
clf2 = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1)
clf3 = SVC(gamma=100.0, C=1.0)
clf4 = KNeighborsClassifier()
clf5 = AdaBoostClassifier()
clf6 = GradientBoostingClassifier()
```

```python
# label encode the target variable
le = LabelEncoder()
y = le.fit_transform(y)

# Code for training machine learning model
for clf in (clf1, clf2, clf3, clf4, clf5, clf6):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

```python
# Save the trained model and label encoder
with open('model.pkl', 'wb') as f:
    pickle.dump((clf6, le), f)
```
