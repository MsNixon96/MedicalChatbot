# Disease Prediction based on Symtoms

```python
#Importing Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn import tree
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score, precision_score
```

# Train the Models using Selected Features

```python
#Reading the training .csv file
df=pd.read_csv("data/Training.csv")
DF= pd.read_csv('data/Training.csv', index_col='prognosis')
#Replace the values in the imported file by pandas by the inbuilt function replace in pandas.

df = df.drop('Unnamed: 133', axis=1)
df.head()
#DF.head()
```

```python
#X= df[l1]
X = df.iloc[:,:-1]
print(type(X))
print(X)
```

```python
y = df[["prognosis"]]
y = np.ravel(y)

# label encode the target variable
le = LabelEncoder()
y = le.fit_transform(y)

print(type(y))
print(y)
```

```python
#Reading the  testing.csv file
tr=pd.read_csv("data/Testing.csv")

tr.head()
```

```python
X_test_csv= tr.iloc[:,:-1]

y_test_csv = tr[["prognosis"]]
y_test_csv = np.ravel(y_test_csv)
y_test_csv = le.transform(y_test_csv)
print(X_test_csv)
```

```python
# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, train_size = 0.70, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

**To select the top 95 features, we used RFE with Random Forest as an estimator**

```python
rfc = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1)
n_features = 95
rfe = RFE(estimator=rfc, n_features_to_select=n_features)
X_train_rfe = rfe.fit_transform(X_train, y_train)
X_test_rfe = rfe.transform(X_test)
X_test_csv_rfe = rfe.transform(X_test_csv)
```

```python
# Get the names of the selected features
selected_feature_names = df.columns[:-1][rfe.support_]
selected_feature_names = list(selected_feature_names)
selected_feature_names = [feature.replace('_', ' ') for feature in selected_feature_names]
print("Selected features:", selected_feature_names)
```

**To build the precision of the model, we utilized three distinctive algorithms which are as per the following**
* Decision Tree algorithm
* Random Forest algorithm
* KNearestNeighbour algorithm
* Naive Bayes algorithm

# Decision Tree Algorithm

```python
from sklearn import tree

clf3 = tree.DecisionTreeClassifier(max_depth=37) 
clf3 = clf3.fit(X_train_rfe,y_train)

y_pred=clf3.predict(X_test_rfe)
print("Decision Tree")
print("Accuracy of Training.csv Split Data")
print("Accuracy score:", accuracy_score(y_test, y_pred))
print("Number of correct predictions:", accuracy_score(y_test, y_pred,normalize=False))
print("Confusion matrix")
conf_matrix=confusion_matrix(y_test,y_pred)
print(conf_matrix)
```

```python
y_pred=clf3.predict(X_test_csv_rfe)
print("Decision Tree")
print("Accuracy of Testing.csv")
print("Accuracy score:", accuracy_score(y_test_csv, y_pred))
print("Number of correct predictions:", accuracy_score(y_test_csv, y_pred, normalize=False))
print("Precision score:", precision_score(y_test_csv, y_pred, average='weighted', zero_division=1))
print("Confusion matrix")
conf_matrix=confusion_matrix(y_test_csv,y_pred)
print(conf_matrix)
```

# Random Forest Algorithm

```python
from sklearn.ensemble import RandomForestClassifier
clf4 = RandomForestClassifier(n_estimators=100)
clf4 = clf4.fit(X_train_rfe,y_train)

# calculating accuracy 
y_pred=clf4.predict(X_test_rfe)
print("Random Forest")
print("Accuracy of Training.csv Split Data")
print("Accuracy score:", accuracy_score(y_test, y_pred))
print("Number of correct predictions:", accuracy_score(y_test, y_pred,normalize=False))
print("Confusion matrix")
conf_matrix=confusion_matrix(y_test,y_pred)
print(conf_matrix)
```

```python
y_pred=clf4.predict(X_test_csv_rfe)
print("Random Forest")
print("Accuracy of Testing.csv")
print("Accuracy score:", accuracy_score(y_test_csv, y_pred))
print("Number of correct predictions:", accuracy_score(y_test_csv, y_pred, normalize=False))
print("Precision score:", precision_score(y_test_csv, y_pred, average='weighted', zero_division=1))
print("Confusion matrix")
conf_matrix=confusion_matrix(y_test_csv,y_pred)
print(conf_matrix)
```

# KNearestNeighbour Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
knn=knn.fit(X_train_rfe,y_train)

y_pred=knn.predict(X_test_rfe)
print("kNearest Neighbour")
print("Accuracy of Training.csv Split Data")
print("Accuracy score:", accuracy_score(y_test, y_pred))
print("Number of correct predictions:", accuracy_score(y_test, y_pred,normalize=False))
print("Confusion matrix")
conf_matrix=confusion_matrix(y_test,y_pred)
print(conf_matrix)
```

```python
y_pred=knn.predict(X_test_csv_rfe)
print("kNearest Neighbour")
print("Accuracy of Testing.csv")
print("Accuracy score:", accuracy_score(y_test_csv, y_pred))
print("Number of correct predictions:", accuracy_score(y_test_csv, y_pred, normalize=False))
print("Precision score:", precision_score(y_test_csv, y_pred, average='weighted', zero_division=1))
print("Confusion matrix")
conf_matrix=confusion_matrix(y_test_csv,y_pred)
print(conf_matrix)
```

# Naive Bayes Algorithm

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb=gnb.fit(X_train_rfe,y_train)


y_pred=gnb.predict(X_test_rfe)
print("Naive Bayes")
print("Accuracy of Training.csv Split Data")
print("Accuracy score:", accuracy_score(y_test, y_pred))
print("Number of correct predictions:", accuracy_score(y_test, y_pred,normalize=False))
print("Confusion matrix")
conf_matrix=confusion_matrix(y_test,y_pred)
print(conf_matrix)
```

```python
y_pred=gnb.predict(X_test_csv_rfe)
print("Naive Bayes")
print("Accuracy of Testing.csv")
print("Accuracy score:", accuracy_score(y_test_csv, y_pred))
print("Number of correct predictions:", accuracy_score(y_test_csv, y_pred, normalize=False))
print("Precision score:", precision_score(y_test_csv, y_pred, average='weighted', zero_division=1))

print("Confusion matrix")
conf_matrix=confusion_matrix(y_test_csv,y_pred)
print(conf_matrix)
```
