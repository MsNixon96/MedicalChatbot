```python
import pickle
import pandas as pd
import numpy as np
```

```python
# load pickled data
with open("model.pkl", "rb") as f:
    model = pickle.load(f)

with open("rfe_model.pickle", "rb") as f:
    rfe = pickle.load(f)

with open("le.pkl", "rb") as f:
    le = pickle.load(f)
```

```python
#Reading the training.csv and testing.csv files
df=pd.read_csv("data/Training.csv")
DF= pd.read_csv('data/Testing.csv')
```

```python
df = df.drop('Unnamed: 133', axis=1)
df.columns = df.columns.str.replace('_', ' ')
df.head()
```

```python
DF.columns = DF.columns.str.replace('_', ' ')
DF.head()
```

```python
feature_names = df.iloc[:,:-1].columns.tolist()
selected_features_indices = rfe.support_
selected_features = [feature_names[i] for i in range(len(feature_names)) if selected_features_indices[i]]
# Print the selected feature names
print("Selected Features:", selected_features)
```

```python
X_test = DF.iloc[:,:-1]
X_test = rfe.transform(X_test)

print(X_test)
```

```python
# Code to test samples

sampleNum = 11

print("Sample {} has the following symptoms:".format(sampleNum))
for i in range(len(selected_features)):
    if X_test[sampleNum][i] == 1:
        print(selected_features[i])
        
print()       
y_pred = model.predict([X_test[sampleNum]])
y_pred = le.inverse_transform(y_pred)
print("The disease prediction for Sample {} is {}".format(sampleNum, y_pred[0])) 
```
